**Extract Certification Title**
- Extract certification titles from  list of dictionaries 
- Append each title to a certification list
- Return certification list

````python
def extractCertificationTitle(certification):
  eval_object = ast.literal_eval(certification)
  certificationList = []
  for item in eval_object:
    if item["title"]:
      certificationList.append(item["title"])
  return certificationList
````

# Feature Extraction

*Stopwords* are a common feature in natural language processing and are frequently used in text analysis tasks. 

Stopwords are words that are commonly used in a language, but they are often ignored when analyzing text because they do not contain significant meaning or value.

example of stop words in english
```
stop_words = ['a', 'an', 'and', 'are', 'as', 'at', 'be', 'by', 'for', 'from', 'has', 'he', 'in', 'is', 'it', 'its', 'of', 'on', 'that', 'the', 'to', 'was', 'were', 'with']

```



````python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words="english")
````

**CountVectorizer** is a popular method for vectorizing text data, which converts a collection of text documents to a matrix of token counts. 



The fit_transform() method of CountVectorizer performs two essential operations:

- **fitting:** it fits the CountVectorizer on the input text data to build a vocabulary, which is a list of all the unique words (tokens) in the text corpus. It also assigns an index to each token in the vocabulary.

- **transforming:** it transforms the input text data into a matrix of token counts, where each row represents a document, and each column represents a unique token in the vocabulary. The matrix stores the frequency of each token in each document.

````python
vectors = cv.fit_transform(df["tags"]).toarray()
````

# Cosine Similarity

````python
from sklearn.metrics.pairwise import cosine_similarity
````

  - The **cosine_similarity()** function calculates the cosine similarity between pairs of vectors in the input matrix vectors. 
  - Cosine similarity is a metric that measures the cosine of the angle between two non-zero vectors in an n-dimensional space. 

  - It ranges between -1 and 1, where 1 indicates that the vectors are identical, and 0 indicates that they are orthogonal (i.e., perpendicular), and -1 indicates that they are diametrically opposed.

  - The output of cosine_similarity(vectors) is a symmetric matrix of shape (n_samples, n_samples), where n_samples is the number of vectors in vectors.

  - The diagonal elements of the matrix are all 1, indicating that each vector is perfectly similar to itself. 

  - The off-diagonal elements represent the cosine similarity between pairs of vectors, where higher values indicate greater similarity.

````python
similarity = cosine_similarity(vectors)
````

````python
class AlumniRecommendationModel:
  def __init__(self, df_pkl, dataDict_pkl ,similarity_pkl):
    self.df = pickle.load(df_pkl)
    self.similarity = pickle.load(similarity_pkl)
    self.dataDict = pickle.load(dataDict_pkl)

  def recommendProfile(self, profileIndex):
    similarProfiles = list(enumerate(self.similarity[profileIndex]))
    sortedSimilarProfiles = sorted(similarProfiles,reverse=True,key = lambda x:x[1])
    top5SimilarProfiles = sortedSimilarProfiles[0:6]
    top5SimilarProfiles
    for i in top5SimilarProfiles:
      print()
      for key in self.dataDict[i[0]]:
        print(f"{key}: {self.dataDict[i[0]][key]}")
      print("-"*100)

  def recommendProfileById(self,id):
    index = self.df[self.df["_id"]=="63f1e38f0666421fed18d222"].index[0]
    return self.recommendProfile(index)
````
